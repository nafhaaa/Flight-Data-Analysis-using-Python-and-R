```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
data_cleaned= pd.read_csv("cleaned_data.csv")
data_plane= pd.read_csv("plane-data.csv")
```

# Take necessary columns in the data_plane and rename the columns based on the data_cleaned dataset

```python
data_plane["manufactured_year"]=data_plane["year"]
data_plane["TailNum"] = data_plane["tailnum"]
```

```python
data_plane=data_plane[["manufactured_year" , "TailNum"]]
```

# Remove rows which has no years under manufactured column

```python
data_plane=data_plane.dropna()
```

# Merge the data_cleaned and data_plane on the tailnum column

```python
merged_data= pd.merge(data_cleaned , data_plane, on="TailNum")
```

```python
merged_data.head()
```

```python
list[merged_data.columns]
```

```python
merged_data.info()
```

```python
#remove the none values and 0 values in the manufactured_column
merged_data=merged_data[(merged_data["manufactured_year"]!= "0000") & (merged_data["manufactured_year"]!= "None")]
```

```python
merged_data
```

```python
pd.isnull(merged_data).sum(axis=0)
```

```python
#since manufactured_year is an object we change it to integer
def clean(changetype):
    changetype= int(changetype)
    return changetype
```

```python
merged_data["Year"]=merged_data["Year"].apply(clean)
```

```python
merged_data["manufactured_year"]=merged_data["manufactured_year"].apply(clean)
```

```python
#add a column which includes the age of plane
```

```python
merged_data["age_of_plane"]=merged_data["Year"]-merged_data["manufactured_year"]
```

```python
merged_data.head()
```

```python
#add a column for the sum of delays
merged_data["sum_of_delays"]=merged_data["DepDelay"]+ merged_data["ArrDelay"]
```

```python
merged_data=merged_data.loc[:, ["Year" ,"manufactured_year" , "age_of_plane" , "sum_of_delays" ]]
```

```python
merged_data
```

```python
#group data by age of plane
delay_by_ageofplane=merged_data.groupby("age_of_plane")["sum_of_delays"].mean().reset_index()
```

```python
delay_by_ageofplane["average_delay_by_age"]=delay_by_ageofplane["sum_of_delays"]
```

```python
#considering age of plane only above 0 
delay_by_ageofplane=delay_by_ageofplane[delay_by_ageofplane["age_of_plane"]>0]
```

```python
delay_by_ageofplane[["age_of_plane" , "average_delay_by_age"]]
```

```python
sns.lmplot(data=delay_by_ageofplane , x="age_of_plane" , y="average_delay_by_age")
```

```python
#check for the correlation
delay_by_ageofplane["age_of_plane"].corr(delay_by_ageofplane["average_delay_by_age"])
```

```python
#there is a weak correlation with a correaltion co-efficient of 0.0479
```

```python
# A LINE plot between the age of plane and the average delay by age
```

```python
sns.lineplot(data=delay_by_ageofplane , x ="age_of_plane" , y="average_delay_by_age")
```
