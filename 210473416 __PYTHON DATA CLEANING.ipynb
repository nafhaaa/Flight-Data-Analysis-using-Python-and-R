# Data cleaning and wrangling


```python
import pandas as pd
import numpy as np
```

# Merging the 2006 and 2007 dataset

```python
#uploading the datasets
dataset_2006=pd.read_csv("2006.csv")
dataset_2007=pd.read_csv("2007.csv")
```

```python
combined_dataset=pd.concat([dataset_2006,dataset_2007])
combined_dataset.head()
```

# Checking for null values 

```python
#checks if there are any null values
combined_dataset.isnull().values.any()
```

```python
#checks the number of  null values in each coloumn
pd.isnull(combined_dataset).sum(axis=0)
```

# Remove the rows which has null values

```python
#98% of the rows under the cancellation column is empty. Therefore we remove the cancellation coloumn .
combined_dataset=combined_dataset.drop(columns="CancellationCode")
```

```python
combined_dataset.head()
```

```python
#remove rows which has null values
combined_dataset.dropna()
```

# Removing time outliers

```python
print(combined_dataset["DepTime"].max())
print(combined_dataset["CRSDepTime"].max())
print(combined_dataset["ArrTime"].max())
print(combined_dataset["CRSArrTime"].max())
```

```python
#The maximum value a 24 hour clock has is 2359 so anything beyond that is not possible . 
# Filter dataframe without values more than 2359 in the Time columns
combined_dataset=combined_dataset[(combined_dataset["DepTime"]<2400) &(combined_dataset["ArrTime"]<2400) & (combined_dataset["CRSArrTime"]<2400) ]
```

```python
#re-confirm the maximum value of the time
print(combined_dataset["DepTime"].max())
print(combined_dataset["CRSDepTime"].max())
print(combined_dataset["ArrTime"].max())
print(combined_dataset["CRSArrTime"].max())
```

```python
#check if the time is greater than zero
print(combined_dataset["DepTime"].min())
print(combined_dataset["CRSDepTime"].min())
print(combined_dataset["ArrTime"].min())
print(combined_dataset["CRSArrTime"].min())
```
